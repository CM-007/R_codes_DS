# **Pre-Processing**

As the data is not transaction formation we are using transaction Encoder

```python
df = pd.get_dummies(titanic)
df.head()
```

# **Apriori Alogorithum**

```python
frequent_item_set = apriori(df,min_support=0.1,use_colnames=True)
frequent_item_set
```

```python
rules = association_rules(frequent_item_set,metric='lift',min_threshold=0.9)
# min_threshold = confidance
rules
rules.sort_values('lift',ascending=False)[:20]
```

```python
rules[rules.lift>1]
```
